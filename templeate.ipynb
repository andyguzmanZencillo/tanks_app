```python
import os

# Función para crear la estructura de carpetas y archivos
def crear_estructura_proyecto(path, nombre):
    # Ruta principal del proyecto
    ruta_principal = os.path.join(path, nombre)
    
    # Crear la carpeta principal
    if not os.path.exists(ruta_principal):
        os.mkdir(ruta_principal)
        print(f'Carpeta principal "{ruta_principal}" creada.')
    
    # Crear subcarpeta "views" y los archivos correspondientes
    ruta_views = os.path.join(ruta_principal, 'views')
    os.mkdir(ruta_views)
    print(f'Subcarpeta "{ruta_views}" creada.')
    
    archivos_views = ['nombre_body.dart', 'nombre_page.dart', 'nombre_view.dart']
    for archivo in archivos_views:
        ruta_archivo = os.path.join(ruta_views, archivo)
        with open(ruta_archivo, 'w') as file:
            file.write(f'// {archivo} contenido')
            print(f'Archivo Dart creado en "{ruta_archivo}".')
    
    # Crear subcarpeta "widgets"
    ruta_widgets = os.path.join(ruta_principal, 'widgets')
    os.mkdir(ruta_widgets)
    print(f'Subcarpeta "{ruta_widgets}" creada.')
    
    # Crear subcarpeta "cubit"
    ruta_cubit = os.path.join(ruta_principal, 'cubit')
    os.mkdir(ruta_cubit)
    print(f'Subcarpeta "{ruta_cubit}" creada.')

# Definir el path donde se creará la estructura y el nombre del proyecto
path = 'lib/features'  # Cambia esta ruta por la que desees
nombre_proyecto = 'nombre'

# Crear la estructura de carpetas y archivos
crear_estructura_proyecto(path, nombre_proyecto)
```

```python
import os

# Función para crear la estructura de carpetas y archivos con código Dart
def crear_estructura_proyecto(path, nombre):
    # Ruta principal del proyecto
    ruta_principal = os.path.join(path, nombre)
    
    # Crear la carpeta principal
    if not os.path.exists(ruta_principal):
        os.mkdir(ruta_principal)
        print(f'Carpeta principal "{ruta_principal}" creada.')
    
    # Crear subcarpeta "views"
    ruta_views = os.path.join(ruta_principal, 'views')
    os.mkdir(ruta_views)
    print(f'Subcarpeta "{ruta_views}" creada.')

    # Crear archivos con las clases Dart basadas en el nombre del proyecto
    archivos_views = [
        {
            'nombre': f'{nombre}_body.dart',
            'contenido': f"""import 'package:flutter/material.dart';
import '{nombre}_view.dart';

class {nombre.capitalize().replace("_", " ").title().replace(" ", "")}Body extends StatelessWidget {{
  const {nombre.capitalize().replace("_", " ").title().replace(" ", "")}Body({{super.key}});

  @override
  Widget build(BuildContext context) {{
    return const Placeholder();   
  }}
}}"""
        },
        {
            'nombre': f'{nombre}_view.dart',
            'contenido': f"""import 'package:flutter/material.dart';
import '{nombre}_body.dart';

class {nombre.capitalize().replace("_", " ").title().replace(" ", "")}View extends StatelessWidget {{
  const {nombre.capitalize().replace("_", " ").title().replace(" ", "")}View({{super.key}});

  @override
  Widget build(BuildContext context) {{
    return const {nombre.capitalize().replace("_", " ").title().replace(" ", "")}Body();  // Contiene la clase {nombre.capitalize()}Body
  }}
}}"""
        },
        {
            'nombre': f'{nombre}_page.dart',
            'contenido': f"""import 'package:flutter/material.dart';
import '{nombre}_view.dart';

class {nombre.capitalize().replace("_", " ").title().replace(" ", "")}Page extends StatelessWidget {{
  const {nombre.capitalize().replace("_", " ").title().replace(" ", "")}Page({{super.key}});

  @override
  Widget build(BuildContext context) {{
    return const {nombre.capitalize().replace("_", " ").title().replace(" ", "")}View();  // Contiene la clase {nombre.capitalize()}View
  }}
}}"""
        }
    ]

    # Crear los archivos Dart en la carpeta "views"
    for archivo in archivos_views:
        ruta_archivo = os.path.join(ruta_views, archivo['nombre'])
        with open(ruta_archivo, 'w') as file:
            file.write(archivo['contenido'])
            print(f'Archivo Dart creado en "{ruta_archivo}".')
    
    # Crear subcarpeta "widgets"
    ruta_widgets = os.path.join(ruta_principal, 'widgets')
    os.mkdir(ruta_widgets)
    print(f'Subcarpeta "{ruta_widgets}" creada.')
    
    # Crear subcarpeta "cubit"
    ruta_cubit = os.path.join(ruta_principal, 'cubit')
    os.mkdir(ruta_cubit)
    print(f'Subcarpeta "{ruta_cubit}" creada.')

# Definir el path donde se creará la estructura y el nombre del proyecto
path = 'lib/features'  # Cambia esta ruta por la que desees
nombre_proyecto = 'sign_up'  # El nombre del proyecto o clase

# Crear la estructura de carpetas y archivos
crear_estructura_proyecto(path, nombre_proyecto)
```
